# 3D Greeks Graphs - Static

### Import visualizer and create an Option object

```python
import visualizer as vis
opt = vis.Option()
```

### The greeks function can be used to show graphs of 2D or 3D graphs (defaulting to 2D). The default arguments for 3D displays Long Call Delta by Time to Expiration and Strike

```python
opt.greeks(graphtype='3D')
```

### Option type and direction can be specified

```python
opt.greeks(graphtype='3D', option='put', direction='short')
```

### The greek parameter can be used to select the option greek or price

### The 3D static graphs take any of the Matplotlib colormaps, specified as colorscheme 

```python
opt.greeks(graphtype='3D', greek='price', colorscheme='inferno')
```

### The axis parameter can be used to display the greek value against volatility and Time to Expiration 

```python
opt.greeks(graphtype='3D', greek='price', axis='vol', option='put') 
```

```python
opt.greeks(graphtype='3D', greek='gamma', direction='short')
```

### The ATM level can be specified with the S0 parameter

```python
opt.greeks(graphtype='3D', greek='vega', S0=50)
```

### The underlying option values can be modified - in this case the interest rate

```python
opt.greeks(graphtype='3D', greek='vomma', r=0.01, direction='short')
```

### And here the level of volatility

```python
opt.greeks(graphtype='3D', greek='vanna', sigma=0.4)
```

```python
opt.greeks(graphtype='3D', greek='theta', option='put', colorscheme='seismic')
```

```python
opt.greeks(graphtype='3D', greek='rho', interactive=False, direction='long')
```

### The chart size can be specified by providing a tuple to the size parameter

```python
opt.greeks(graphtype='3D', greek='zomma', direction='short', size=(10, 8))
```

```python
opt.greeks(graphtype='3D', greek='speed', colorscheme='viridis')
```

### The color transparency can be specified with the colorintensity parameter

```python
opt.greeks(graphtype='3D', greek='color', direction='short', colorintensity=0.8)
```

```python
opt.greeks(graphtype='3D', greek='ultima', direction='short')
```

```python
opt.greeks(graphtype='3D', greek='vega bleed', direction='short')
```

```python
opt.greeks(graphtype='3D', greek='charm', option='put')
```
