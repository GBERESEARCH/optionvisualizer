# 2D Greeks Graphs

### Import pricer and create an Option object

```python
import visualizer as vis
opt = vis.Option()
```

### The function default arguments displays a graph of Long 3M Call Delta vs Time with Strikes at 90%, ATM and 110%

```python
opt.greeks_graphs_2D()
```

### The x and y axis arguments are referenced using x_plot and y_plot, here Delta vs Price

```python
opt.greeks_graphs_2D(x_plot='delta', y_plot='price', G1=100, G2=100, G3=100, T1=0.05, T2=0.15, T3=0.25)
```

### Strikes are referenced as G1, G2 & G3 and Direction as long or short, here Delta vs Volatility

```python
opt.greeks_graphs_2D(x_plot='delta', y_plot='vol', G1=80, G2=90, G3=100, direction='short')
```

### Option can be specified as call or put, here Delta vs Time 

```python
opt.greeks_graphs_2D(x_plot='delta', y_plot='time', option='put', G1=105, G2=110, G3=115, direction='short')
```
