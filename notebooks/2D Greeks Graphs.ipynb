# 2D Greeks Graphs

### Import visualizer and create an Option object

```python
import optionvisualizer.visualizer as vis
opt = vis.Option()
```

### The greeks function can be used to show graphs of 2D or 3D graphs (defaulting to 2D). The default arguments displays Long 3M Call Delta vs Time with Strikes at 90%, ATM and 110%

```python
opt.greeks(graphtype='2D')
```

### The x and y axis arguments are referenced using x_plot and y_plot, here Delta vs Price

```python
opt.greeks(x_plot='price', y_plot='delta', G1=100, G2=100, G3=100, T1=0.05, T2=0.15, T3=0.25)
```

## Alternatively use the 'visualize' method which can show risk or payoff graphs using the 'risk' argument

### Strikes are referenced as G1, G2 & G3 and Direction as long or short, here Delta vs Volatility

```python
opt.visualize(risk=True, x_plot='vol', y_plot='delta', G1=80, G2=90, G3=100, direction='short')
```

### Option can be specified as call or put, here Delta vs Time 

```python
opt.visualize(risk=True, x_plot='time', y_plot='delta', option='put', G1=105, G2=110, G3=115, direction='short')
```
