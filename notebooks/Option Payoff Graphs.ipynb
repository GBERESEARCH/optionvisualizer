# Option Payoff Graphs

### Import pricer and create an Option object 

```python
import visualizer as vis
opt = vis.Option()
```

### Payoffs can be referenced directly by name and come with default parameters so can be run without arguments 

```python
opt.call()
```

### Or you can be explicit about any or all arguments

```python
opt.call(S=90, K=95, T=0.75, r=0.05, q=0, sigma=0.3, direction='short', value=True)
```

```python
opt.put()
```

```python
opt.stock()
```

```python
opt.forward()
```

### Alternatively each payoff can be accessed from the payoffs function via the combo payoff argument

```python
opt.payoffs(combo_payoff='collar')
```

```python
opt.payoffs(combo_payoff='spread', direction='long', option='put')
```

```python
opt.payoffs(combo_payoff='backspread', ratio=3, value=True)
```

```python
opt.payoffs(combo_payoff='ratio vertical spread', ratio=3, option='put')
```

```python
opt.payoffs(combo_payoff='straddle')
```

```python
opt.payoffs(combo_payoff='strangle', S=50, K1=45, K2=55, direction='short')
```

```python
opt.payoffs(combo_payoff='butterfly', value=True)
```

```python
opt.payoffs(combo_payoff='christmas tree', value=True, direction='short')
```

```python
opt.payoffs(combo_payoff='condor', option='put')
```

```python
opt.payoffs(combo_payoff='iron butterfly')
```

```python
opt.payoffs(combo_payoff='iron condor', value=True, direction='short')
```
